```python
# Import relevant modules
import random
import pickle as pkl
import numpy as np
import os
print(os.getcwd())

# Define the amount of each type of rhythm
N_num = 25477
AFIB_num = 11274
P_num = 7848
B_num = 2853
SBR_num = 1776
AFL_num = 1306
T_num = 1194
label_names = ['N', 'AFIB', 'P', 'B', 'SBR', 'AFL', 'T']
label_amounts = [N_num, AFIB_num, P_num, B_num, SBR_num, AFL_num, T_num]
print(label_names)
print(label_amounts)
```

```python
# Cycle through rhythm labels we want to sort
for i in range(len(label_names)):
    seg_counter = 0
    # Cycle through all patients
    for j in range(48):
        # Load in all segments and labels of the patient's ecg
        with open('3 Beat Segments/{}_segs.pkl'.format(j), 'rb') as f:
            segments = pkl.load(f)
        #with open('3 Beat Segments Noisy/{}_segs.pkl'.format(j), 'rb') as f:
        #    noisy_segments = pkl.load(f)
        with open('3 Beat Labels/{}_labels.pkl'.format(j), 'rb') as f:
            labels = pkl.load(f)
        # Now cycle through each segment
        for k in range(len(segments)):
            # and sort into folder if label matches 
            if labels[k] == label_names[i]:
                with open('Sorted Segs/{}/seg_{}.pkl'.format(label_names[i], seg_counter), 'wb') as f:
                    pkl.dump(segments[k], f)
                #with open('Sorted Segs Noisy/{}/seg_{}.pkl'.format(label_names[i], seg_counter), 'wb') as f:
                #    pkl.dump(noisy_segments[k], f)
                seg_counter += 1
            
```

```python

for i in range(len(label_names)):
    print(i)
    # Make array of all file numbers
    index_array = []
    for j in range(label_amounts[i]):
        index_array.append(j)
    # Shuffle this array of all file numbers of a particular category
    random.shuffle(index_array)
    short_array = []
    # Choose the first 1180 of these shuffled indices
    for k in range(1180):
        short_array.append(index_array[k])
    # Now cycle through the 1180 long array and copy over data to new folder
    count = 0
    for file_num in short_array:
        print(file_num)
        with open('Sorted Segs/{}/seg_{}.pkl'.format(label_names[i], file_num), 'rb') as f:
            seg = pkl.load(f)
        with open('Short Sorted Segs/{}/seg_{}.pkl'.format(label_names[i], count), 'wb') as f:
            pkl.dump(seg, f)
        with open('Sorted Segs Noisy/{}/seg_{}.pkl'.format(label_names[i], file_num), 'rb') as f:
            seg = pkl.load(f)
        with open('Short Sorted Segs Noisy/{}/seg_{}.pkl'.format(label_names[i], count), 'wb') as f:
            pkl.dump(seg, f)
        count += 1
    with open('Short Sorted Segs/{}_file_nums.pkl'.format(label_names[i]), 'wb') as f:
        pkl.dump(short_array, f)
    with open('Short Sorted Segs Noisy/{}_file_nums.pkl'.format(label_names[i]), 'wb') as f:
        pkl.dump(short_array, f)
    print('done')
```

```python
# move sorted segments into one folder and make corresponding label array
# also do same with noisy data (were gonna try and train on both noisy and denoised but ran out of time)

count = 0
# make array of all labels
label_array = []
for i in range(len(label_names)):
    print(i)
    file_num = 0
    for j in range(1180):
        with open('Short Sorted Segs/{}/seg_{}.pkl'.format(label_names[i], file_num), 'rb') as f:
            seg = pkl.load(f)
        with open('3 Beat Train Data/seg_{}.pkl'.format(count), 'wb') as f:
            pkl.dump(seg, f)
        with open('Short Sorted Segs Noisy/{}/seg_{}.pkl'.format(label_names[i], file_num), 'rb') as f:
            seg = pkl.load(f)
        with open('3 Beat Train Data Noisy/seg_{}.pkl'.format(count), 'wb') as f:
            pkl.dump(seg, f)
        label_array.append(label_names[i])
        count += 1
        file_num += 1
    print('done')
```

```python
# save label
with open('CNN_3_beat_labels.pkl', 'wb') as f:
    pkl.dump(label_array, f)
print(label_array)
print(len(label_array))
```

```python
# one hot the boys
from keras.utils import to_categorical
import pickle
Labels = []
with open('CNN_3_beat_labels.pkl', 'rb') as f:
    Labels = pickle.load(f)
    
print(Labels[0])
# Convert to numbers corresponding to character label
label_cat = [0 if b=='N'
            else 1 if b=='AFIB'
            else 2 if b=='P'
            else 3 if b=='B'
            else 4 if b=='SBR'
            else 5 if b=='AFL'
            else 6 if b=='T'
            else 7 for b in Labels]
with open('CNN_3_Beat_label_cat.pkl', 'wb') as f:
    pickle.dump(label_cat, f)
    
print(label_cat[0])
# Convert to one hot encoding
# representation of categorical variables as binary vectors
label_OHE = to_categorical(label_cat)
with open('CNN_3_Beat_label_OHE.pkl', 'wb') as f:
    pickle.dump(label_OHE,f)
    
print(label_OHE[0])
```
